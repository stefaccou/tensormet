```python
from tucker_tensor import TuckerDecomposition
dataset = "fineweb-es"
dims = 1000
rank = 150
iterations = 750
```

```python
sc_tucker = TuckerDecomposition.load_from_disk(
    dataset=dataset,
    method="sc",
    name=None,
    divergence="kl",
    dims=dims,
    rank=rank,
    iterations=iterations,
    map_location="cpu"
)
```

```python
# we visualize the slice for the verb "spelen"
# test_tuple = ("spelen", "kind", "bal")
test_tuple = ("jugar", "chico", "juego")
role = "verb"
sc_tucker.visualize_slice(test_tuple, role=role, normalize=True, method="slice")
```

```python
top_activations = sc_tucker.retrieve_highest_activations(test_tuple, role="verb", method="weighted_tuple", top_k=10)
for (id_1, id_2), value in top_activations:
    top_words_dim1 = sc_tucker.get_top_words_for_dimension(role="subject", dim_index=id_1, top_k=5)
    top_words_dim2 = sc_tucker.get_top_words_for_dimension(role="object", dim_index=id_2, top_k=5)
    print(f"{value}: {id_1}, {id_2}")
    print(f"\t{id_1}:{[word for word, score in top_words_dim1]}")
    print(f"\t{id_2}:{[word for word, score in top_words_dim2]}")
```
