Parsing the e-lex to be easily checkeable

```python
import xml.etree.ElementTree as ET
from html.entities import name2codepoint
from utils import DATA_DIR

data = DATA_DIR / "lexica" / "e-Lex-1.1.xml"
parser = ET.XMLParser()
parser.entity.update({name: chr(cp) for name, cp in name2codepoint.items()})
tree = ET.parse(data, parser=parser)
root = tree.getroot()
verbs_to_forms = {}
for entry in root.findall(".//entry"):
    lemma = entry.find("lem")
    wordform_list = []
    for word in entry.findall("wordf"):
        if word.find("pos").text.startswith("WW"):
            wordform = word.find("orth")
            wordform_list.append(wordform.text)
    if wordform_list:
        verbs_to_forms[lemma.text] = wordform_list

verbs_to_forms
```
